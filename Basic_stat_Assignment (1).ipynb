```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
```

```python
#Function to identify outliers
def identify_outliers(column):
  Q1=column.quantile(0.25)
  Q3=column.quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-1.5*IQR
  upper_bound=Q3+1.5*IQR
  outliers=column[(column<lower_bound)|(column>upper_bound)]
  return outliers

#Plot histograms
for col in numerical_col:
  plt.figure(figsize=(8,6))
  sns.histplot(data[col],kde=True,color='skyblue',edgecolor='black')
  skewness_value=skew(data[col])
  outliers=identify_outliers(data[col])
  plt.scatter(outliers,[0]* len(outliers),color='red',label='outliers',zorder=2)
  plt.title(f'Histogram of {col}\n Skewness: {skewness_value:.2f}',fontsize=14)
  plt.xlabel(col)
  plt.ylabel('Frequency')
  plt.legend()
  plt.show()
```

```python
# plot boxplot

for col in numerical_col:
  plt.figure(figsize=(8,6))
  sns.boxplot(x=data[col],color='skyblue',whis=1.5)

  # calculate IQR
  Q1=data[col].quantile(0.25)
  Q3=data[col].quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-1.5*IQR
  upper_bound=Q3+1.5*IQR

  plt.title(f'Boxplot of {col}\nIQR:{IQR:.2f}| Lower Bound:{lower_bound:2f}|Upper Bound:{upper_bound:2f}',fontsize=14)
  plt.xlabel(col)
  plt.show()
```

```python
categorical_col=data.select_dtypes(include=['object']).columns
print(categorical_col)
```

```python
def plot_cat_distribution(categorical_col):
  for columns in categorical_col:
    plt.figure(figsize=(8,6))
    value_counts=categorical_col[columns].value_counts()
    plt.bar(value_counts.index,value_counts.values,color='skyblue',edgecolor='black')
    plt.title(f'Count Plot of {columns}',fontsize=14)
    plt.xlabel(columns)
    plt.ylabel('Count')
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.tight_layout()
    plt.show()

    #Insights
    total=value_counts.sum()
    percentage=(value_counts/total)*100
    print(f'Insights for {columns}:')
    for category,count in value_counts.items():
      print(f'{category}: {count} occurences ({percentage[category]:.2f}%)')
    print('-'*40)
plot_cat_distribution(data[categorical_col])
```

# Explain Standardization( Z-score Normalization )
Standardization is a technique used in data preprocessing to scale the
values of a dataset such that they have a mean of 0 and a standard deviation of 1.
This process is particularly useful when your dataset contains features with differeent
units or scales, as it ensures that each feature contributes equally to the analysis or
model training.

```python
standardized_data=data.copy()
standardized_data[numerical_col]=(data[numerical_col]-data[numerical_col].mean())/data[numerical_col].std()

print("Original DataFrame:")
print(data)
print("\nStandardized DataFrame:")
print(standardized_data)
```

# Conversion of categorical data into dummy variables

Need to convert categorical data into dummy variables:
1. Machine Learning Models Require Numerical data, One-hot encoding converts categorical variables into binary (0 or 1) columns,making them usable by these algorithm.
2. Preserve Information in Categorical Variables: Each unique category gets its own column in one-hot encoding. This ensures that no information is lost during the encoding process.

```python
# One-hot Encoding
encode=pd.get_dummies(data[categorical_col],drop_first=False)
print("Original Dataset:")
print(data[categorical_col])
print("\nEncoded Dataset:")
print(encode)
```
