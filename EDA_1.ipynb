Detecting the Outliers

```python
#Function to identify outliers
from scipy.stats import skew
def identify_outliers(column):
  Q1=column.quantile(0.25)
  Q3=column.quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-1.5*IQR
  upper_bound=Q3+1.5*IQR
  outliers=column[(column<lower_bound)|(column>upper_bound)]
  return outliers

#Plot histograms
for col in numerical_col:
  plt.figure(figsize=(8,6))
  sns.histplot(df[col],kde=True,color='skyblue',edgecolor='black')
  skewness_value=skew(df[col])
  outliers=identify_outliers(df[col])
  plt.scatter(outliers,[0]* len(outliers),color='red',label='outliers',zorder=2)
  plt.title(f'Histogram of {col}\n Skewness: {skewness_value:.2f}',fontsize=14)
  plt.xlabel(col)
  plt.ylabel('Frequency')
  plt.legend()
  plt.show()
```

**3.Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")

# List of numerical columns to visualize
num_cols = df.select_dtypes(include=['float64']).columns

# Create histograms
plt.figure(figsize=(15, 12))
for i, col in enumerate(num_cols, 1):
    plt.subplot(4, 4, i)
    sns.histplot(df[col].dropna(), bins=30, kde=True)
    plt.title(f"Histogram of {col}")
plt.tight_layout()
plt.show()

# Create boxplots
plt.figure(figsize=(15, 12))
for i, col in enumerate(num_cols, 1):
    plt.subplot(4, 4, i)
    sns.boxplot(x=df[col].dropna())
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

**Key Correlations & Their Potential Implications:**

**1.Moderate Positive Correlations:**

ASTV (Abnormal Short-Term Variability) & NSP (0.42)
:Higher abnormal variability in fetal heart rate is linked with increased NSP values, possibly indicating non-reassuring fetal status.

ALTV (Abnormal Long-Term Variability) & NSP (0.37)
: Similar to ASTV, long-term variability abnormalities are associated with changes in fetal state classification.

DP (Decelerations Percentage) & NSP (0.29)
: Increased deceleration rates may contribute to a worsening fetal state.

**2.Negative Correlations:**
AC (Accelerations) & NSP (-0.32)
:A higher number of accelerations is linked to a more favorable fetal state, supporting the idea that fetal activity can indicate well-being.


**3.Other Notable Correlations**
Width & MSTV (Mean Short-Term Variability) (0.44)
: Suggests that width might be influenced by short-term fluctuations in heart rate.

DL (Light Decelerations) & Width (0.41)
: Indicates that fetal decelerations might impact the overall variability range.

**Possible Medical Interpretation:**
Higher ASTV and ALTV values may indicate fetal distress and potential need for further monitoring.

A decrease in accelerations could suggest an abnormal fetal state, possibly requiring medical intervention.

Deceleration patterns (DP & NSP correlation) further support the importance of monitoring fetal heart rate trends.
