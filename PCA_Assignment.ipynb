```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.cluster import KMeans
```

```python
X = df[columns[1:]].values
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA()
```

```python
pca.fit(X_scaled)
```

```python
explained_variance = pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
plt.title("Scree Plot")
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.show()
```

```python
cumulative_variance = np.cumsum(explained_variance)
# Plot cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.axhline(y=0.95, color='r', linestyle='--')  # Typically, we select components that explain 95% variance
plt.title("Cumulative Explained Variance")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.show()
```

```python
optimal_components = np.where(cumulative_variance >= 0.95)[0][0] + 1  # Adding 1 because Python indexing starts at 0
print(f"Optimal number of components: {optimal_components}")
```

```python
pca_optimal = PCA(n_components=optimal_components)
X_pca = pca_optimal.fit_transform(X_scaled)
```

```python
df_pca = pd.DataFrame(X_pca, columns=[f"PC{i+1}" for i in range(optimal_components)])
# Show the transformed data
print(df_pca.head())
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
kmeans.fit(X_scaled)
# Assign the cluster labels to the dataset
df['Cluster'] = kmeans.labels_
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=df['Cluster'], palette='Set2', s=100, alpha=0.7, edgecolor='k')
plt.title("K-means Clustering (2D Projection)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend(title="Cluster", loc='upper right')
plt.show()
```

```python
silhouette_avg = silhouette_score(X_scaled, df['Cluster'])
print(f"Silhouette Score: {silhouette_avg}")
```

```python
davies_bouldin_avg = davies_bouldin_score(X_scaled, df['Cluster'])
print(f"Davies-Bouldin Index: {davies_bouldin_avg}")
```

1. Summary of Key Findings and Insights

Feature Distribution:

Visualizations (histograms, boxplots, and density plots) revealed variations in the scale and distribution of features, justifying the need for feature standardization.

Some features were skewed, and many showed differences in range and variability.

K-means clustering was applied to both the original standardized data and PCA-transformed data.

Silhouette Score and Daviesâ€“Bouldin Index were used to evaluate clustering quality.

PCA-transformed data often led to better or comparable clustering performance, with more visually separable clusters.



2. Practical Implications of Using PCA and Clustering in Data Analysis

Dimensionality Reduction with PCA:

Helps remove noise and multicollinearity from the data.

Reduces computational complexity, especially for high-dimensional datasets.

Enables effective visualization of high-dimensional data in 2D or 3D.

Clustering (K-means):

Useful for discovering inherent groupings within the data.

Assists in market segmentation, customer profiling, anomaly detection, etc.

Performance can be impacted by the shape, scale, and dimensionality of the input data.



3. Recommendations for When to Use Each Technique

High-Dimensional Data-	Use PCA first to reduce dimensions before clustering.

Correlated Features Present	-Apply PCA to remove redundancy and improve clustering.

Need for Interpretation-	Prefer clustering on the original dataset to retain feature context.

Visualization Required-Use PCA to transform data for 2D or 3D visualization.
